```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import numpy as np 
import pandas as pd 

pd.set_option("display.float_format", lambda x: "%.5f" %x)
```

```python
df_ = pd.read_excel("/kaggle/input/online-retail-ii/online_retail_II.xlsx")
```

```python
df = df_.copy()
df.head()
df.columns
df.isnull().sum()
```

```python
df = df[~df["Invoice"].str.contains("C", na=False)]
df.head()
```

```python
df.describe().T
```

```python
df = df[(df["Quantity"] > 0)]
df.head()
```

```python
df.dropna(inplace=True)
df.describe().T
```

```python
df["TotalPrice"] = df["Quantity"] * df["Price"]
df.head()
```

```python
cltv_c = df.groupby("Customer ID").agg({"Invoice": lambda x: x.nunique(),
                                        "Quantity": lambda x: x.sum(),
                                        "TotalPrice": lambda x: x.sum()})
cltv_c.head()
```

```python
cltv_c.columns = ["total_transaction", "total_unit", "total_price"]
cltv_c.head()
```

## Average Order Value

Average Order Value = total price / total transaction

```python
cltv_c["average_order_value"] = cltv_c["total_price"] / cltv_c["total_transaction"]
cltv_c.head()
```

## Purchase Frequency

Purchase Frequency = Total transaction / total number of customers

```python
total_number_of_customers = cltv_c.shape[0]
cltv_c["purchase_frequency"] = cltv_c["total_transaction"] / cltv_c.shape[0]
cltv_c.head()
```

## Repeat Rate & Churn Rate

```python
repeat_rate = cltv_c[cltv_c["total_transaction"] > 1].shape[0] / cltv_c.shape[0]
print(repeat_rate)
```

```python
churn_rate = 1 - repeat_rate 
print(churn_rate)
```

## Profit Margin


Profit Margin = total price * 0.10

(The company itself determines the constant value of 0.10.)

```python
cltv_c["profit_margin"] = cltv_c["total_price"] * 0.10
cltv_c.head()
```

## Customer value

Customer value = average order value * purchase frequency

```python
cltv_c["customer_value"] = cltv_c["average_order_value"] * cltv_c["purchase_frequency"]
cltv_c.head()
```

## Customer Lifetime Value

CLTV = (customer_value / churn_rate ) * profit_margin

```python
cltv_c["cltv"] = (cltv_c["customer_value"] / churn_rate) * cltv_c["profit_margin"]
cltv_c.head()
```

```python
cltv_c.sort_values(by="cltv", ascending=False)
```

```python
cltv_c.describe().T
```

```python
cltv_c["segment"] = pd.qcut(cltv_c["cltv"], 4, labels=["D", "C", "B", "A"])
cltv_c.head()
```

```python
cltv_c.groupby("segment", observed=True).agg({"count", "mean", "sum"})
```
